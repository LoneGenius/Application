```python
# Print all names under 'Aktiviti Ekonomi'
print(df['Aktiviti Ekonomi\nEconomic Activity'].to_list())
```

```python
df_pivoted = pd.melt(df, id_vars=['Aktiviti Ekonomi\nEconomic Activity'], 
                     value_vars=['2020', '2021', '2022', '2023'], 
                     var_name='Year', value_name='Total')
```

```python
def classify_stem(row):
    stem_fields = [
        "Pertanian\nAgriculture", 
        "Perlombongan dan pengkuarian\nMining and quarrying", 
        "Pembuatan - Minyak dan lemak daripada sayuran & haiwan dan prosesan makanan\nVegetable and animal oils & fats and food processing", 
        "Pembuatan - Minuman dan produk tembakau\nBeverages and tobacco products", 
        "Pembuatan - Produk tekstil, pakaian dan kulit\nTextiles, wearing apparel and leather products", 
        "Pembuatan - Produk kayu, perabot, keluaran kertas dan percetakan\nWood products, furniture, paper products and printing", 
        "Pembuatan - Produk petroleum, kimia, getah dan plastik\nPetroleum, chemical, rubber and plastic products", 
        "Pembuatan - Produk mineral bukan logam, logam asas dan produk logam yang direka\nNon-metallic mineral products, basic metal and fabricated metal products", 
        "Pembuatan - Produk elektrik, elektronik dan optikal\nElectrical, electronic and optical products", 
        "Pembuatan - Peralatan pengangkutan, pembuatan lain dan pembaikan\nTransport equipment, other manufacturing and repair"
    ]
    if any(field in row for field in stem_fields):
        return "STEM"
    else:
        return "Non-STEM"

df_pivoted['STEM Classification'] = df_pivoted['Aktiviti Ekonomi\nEconomic Activity'].apply(classify_stem)
```

```python
df_pivoted.head()
```

```python
# Step 6: Multiply year columns by 1000 to adjust the values
for col in ['Total']:
    df_pivoted[col] = df_pivoted[col] * 1000
```

```python
# Save the pivoted data to a new CSV file
df_pivoted.to_csv("grads_field_employed_DOSM.csv", index=False)
```
