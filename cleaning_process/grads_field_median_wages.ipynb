```python
df = pd.read_excel(grads_median_wages, sheet_name="Jadual 10i - 10l", header=None)
```

```python
df.columns = df.iloc[0]
```

```python
# Step 2: Remove the "Sektor\nSectors" row
df = df[df['Sektor\nSectors'] != 'Sektor\nSectors']
```

```python
# Step 2: Remove the "Sektor\nSectors" row
df = df[df['Sektor\nSectors'] != 'Jumlah \nTotal']
```

```python
df
```

```python
# Keep "Aktiviti Ekonomi" and remove the next 4 columns
df = df[["Sektor\nSectors", "2020r", "2021r", "2022r", 2023]]
```

```python
df
```

```python
df = df.rename(columns={'2020r': '2020', '2021r': '2021', '2022r': '2022', 2023: '2023'})
```

```python
df
```

```python
df_pivoted = pd.melt(df, id_vars=['Sektor\nSectors'], 
                     value_vars=['2020', '2021', '2022', '2023'], 
                     var_name='Year', value_name='Total')
```

```python
# Save the pivoted data to a new CSV file
df_pivoted.to_csv("grads_median_wages_DOSM.csv", index=False)
```
