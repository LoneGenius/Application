```python
df = pd.read_excel(grads_states_median_wages, sheet_name="Jadual 10m - 10n", header=None)
```

```python
df = df.iloc[25:42]
```

```python
df.columns = df.iloc[0]
```

```python
# Step 2: Remove the "Negeri\nState" row
df = df[df['Negeri\nState'] != 'Negeri\nState']
```

```python
# Step 2: Remove the "Negeri\nState" row
df = df[df['Negeri\nState'] != 'Jumlah \nTotal']
```

```python
df
```

```python
# Keep "Negeri\nState" and remove the next 4 columns
df = df[["Negeri\nState", "2020r", "2021r", "2022r", 2023]]
```

```python
df = df.rename(columns={'2020r': '2020', '2021r': '2021', '2022r': '2022', 2023: '2023'})
```

```python
df
```

```python
df_pivoted = pd.melt(df, id_vars=['Negeri\nState'], 
                     value_vars=['2020', '2021', '2022', '2023'], 
                     var_name='Year', value_name='Total')
```

```python
# Save the pivoted data to a new CSV file
df_pivoted.to_csv("grads_states_median_wages_DOSM.csv", index=False)
```
