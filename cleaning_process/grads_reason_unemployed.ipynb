```python
import pandas as pd
```

```python
grads_reason_unemployed= r"C:\Users\Aqilah Maisarah\OneDrive\MDIT\Application\Dataset\Jadual Penerbitan Statistik Siswazah 2023 (1).xlsx"
```

```python
df = pd.read_excel(grads_reason_unemployed, sheet_name="Jadual 8d", header=None)
```

```python
df.head()
```

```python
df = df.iloc[3:]  # Skip the first 4 rows which are metadata and headers
```

```python
df.head()
```

```python
df.columns = df.iloc[0]
```

```python
df=df.iloc[2:8]
```

```python
df
```

```python
# Step 2: Remove the "Sebab Tidak Mencari Pekerjaan\nReasons For Not Seeking Work" row
df = df[df['Sebab Tidak Mencari Pekerjaan\nReasons For Not Seeking Work'] != 'Sebab Tidak Mencari Pekerjaan\nReasons For Not Seeking Work']
```

```python
# Remove the "Jumlah \nTotal" row
df = df[df['Sebab Tidak Mencari Pekerjaan\nReasons For Not Seeking Work'] != 'Jumlah \nTotal']
```

```python
# Keep "Aktiviti Ekonomi" and remove the next 4 columns
df = df[["Sebab Tidak Mencari Pekerjaan\nReasons For Not Seeking Work", "2020r", "2021r", "2022r", 2023]]
```

```python
df = df.rename(columns={'2020r': '2020', '2021r': '2021', '2022r': '2022', 2023: '2023'})
```

```python
df
```

```python
df_pivoted = pd.melt(df, id_vars=['Sebab Tidak Mencari Pekerjaan\nReasons For Not Seeking Work'], 
                     value_vars=['2020', '2021', '2022', '2023'], 
                     var_name='Year', value_name='Total')
```

```python
# Step 6: Multiply year columns by 1000 to adjust the values
for col in ['Total']:
    df_pivoted[col] = df_pivoted[col] * 1000
```

```python
df_pivoted
```

```python
# Save the pivoted data to a new CSV file
df_pivoted.to_csv("grads_reason_unemployed_DOSM.csv", index=False)
```
